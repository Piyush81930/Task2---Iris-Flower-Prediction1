# Best Model Finding

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
Svm_model = cross_val_score(SVC(),df[['sepal_length','sepal_width','petal_length','petal_width']],df.Species_n)
Svm_model
```

```python
randomforest = cross_val_score(RandomForestClassifier(),df[['sepal_length','sepal_width','petal_length','petal_width']],df.Species_n)
randomforest
```

```python
decisiontree = cross_val_score(DecisionTreeClassifier(),df[['sepal_length','sepal_width','petal_length','petal_width']],df.Species_n)
decisiontree
```

```python
logistic = cross_val_score(LogisticRegression(),df[['sepal_length','sepal_width','petal_length','petal_width']],df.Species_n)
logistic
```

```python
linear = cross_val_score(LinearRegression(),df[['sepal_length','sepal_width','petal_length','petal_width']],df.Species_n)
linear
```

According to cross_val_score Logistic Regression Accuracy is Best


# Model Training and Evaluation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1,df2,test_size =0.2,random_state =42)
```

```python
model = LogisticRegression()
model
```

```python
model.fit(X_train,y_train)
```

0 - Iris-setosa
1 - Iris-versicolor
2 - Iris-virginica

```python
model.predict([[5.1,3.5,1.4,0.2]])
```

```python
model.predict([[6.1,7.8,9.1,0.9]])
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_pred,y_test)
cm
```

```python
plt.figure(figsize =(10,7))
sns.heatmap(cm , annot =True)
plt.xlabel("predicted")
plt.ylabel("truth beheind t=this")
```
